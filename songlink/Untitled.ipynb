```python
import librosa
import numpy as np
import pandas as pd
import os
import matplotlib.pyplot as plt
import plotly.graph_objs as go
import plotly.offline as pyo
import seaborn as sns





# 폴더 이름(데이터로 쓸 샘플들 모아놓은 거)
folder_path = "forsea"


try:
    df = pd.read_csv("DF.csv")
except FileNotFoundError:
    df = pd.DataFrame()


for file_name in os.listdir(folder_path):
    file_path = os.path.join(folder_path, file_name)

    # librosa 주로 mp3, wav 파일 분석함ㅇㅇ
    if file_name.endswith(".mp3") or file_name.endswith(".wav"):
        # 음성 파일 불러오기
        signal, sr = librosa.load(file_path, sr=44000)

        # MFCC 추출
        mfcc = librosa.feature.mfcc(y=signal, sr=sr, n_mfcc=20)

        # 스펙트럴 중심주파수 추출
        spectral_centroids = librosa.feature.spectral_centroid(y=signal, sr=sr)

        # 스펙트럴 대역폭 추출
        spectral_bandwidth = librosa.feature.spectral_bandwidth(y=signal, sr=sr)

        # 스펙트럴 콘트라스트 추출
        spectral_contrast = librosa.feature.spectral_contrast(y=signal, sr=sr)

        # 스펙트럴 롤오프 추출
        spectral_rolloff = librosa.feature.spectral_rolloff(y=signal, sr=sr)

        # 평균 주파수
        meanfreq = np.mean(spectral_centroids)

        # 최고 주파수
        maxfreq = np.max(spectral_centroids)

        # 최저 주파수
#         minfreq = np.min(spectral_centroids)


#         mfcc_mean = np.mean(mfcc[:, 1:], axis=1).tolist()
#         spectral_centroids_mean = np.mean(spectral_centroids[spectral_centroids > 0])
#         spectral_bandwidth_mean = np.mean(spectral_bandwidth[spectral_bandwidth > 0])
#         spectral_contrast_mean = np.mean(spectral_contrast[spectral_contrast > 0], axis=0).tolist()
#         spectral_rolloff_mean = np.mean(spectral_rolloff[spectral_rolloff > 0])
#         meanfreq_mean = np.mean(meanfreq[meanfreq > 0])
#         maxfreq_mean = np.mean(maxfreq[maxfreq > 0])
#         minfreq_mean = np.mean(minfreq[minfreq > 0])

#         result = {"mfcc_"+str(i+1): mfcc_mean[i] for i in range(20)}
#         result["spectral_centroids"] = spectral_centroids_mean
#         result["spectral_bandwidth"] = spectral_bandwidth_mean
#         result["spectral_contrast"] = spectral_contrast_mean
#         result["spectral_rolloff"] = spectral_rolloff_mean
#         result["meanfreq"] = meanfreq_mean
#         result["maxfreq"] = maxfreq_mean
#         result["minfreq"] = minfreq_mean
#         result["label"]=1  #남자0 여자1
#         df = df.append(result, ignore_index=True)
        




# 엑셀 파일에 저장
# df.to_csv("DF.csv", index=False)
mfcc_mean = df["mfcc_1"].values[:80]
spectral_centroids_mean = df["spectral_centroids"].values[:150]
spectral_bandwidth_mean = df["spectral_bandwidth"].values[:150]
spectral_contrast_mean = df["spectral_contrast"].values[:150]
spectral_rolloff_mean = df["spectral_rolloff"].values[:150]
meanfreq_mean = df["meanfreq"].values[:150]
maxfreq_mean = df["maxfreq"].values[:150]
minfreq_mean = df["minfreq"].values[:150]

x1 = list(range(-100, 100))
fig = go.Figure()
fig.add_trace(go.Scatter(x=x1, y=spectral_centroids_mean, mode='lines', name='spectral_centroids_mean'))
fig.add_trace(go.Scatter(x=x1, y=spectral_bandwidth_mean, mode='lines', name='spectral_bandwidth_mean'))
fig.add_trace(go.Scatter(x=x1, y=spectral_contrast_mean, mode='lines', name='spectral_contrast_mean'))
fig.add_trace(go.Scatter(x=x1, y=spectral_rolloff_mean, mode='lines', name='spectral_rolloff_mean'))
fig.add_trace(go.Scatter(x=x1, y=meanfreq_mean, mode='lines', name='meanfreq_mean'))
fig.add_trace(go.Scatter(x=x1, y=maxfreq_mean, mode='lines', name='maxfreq_mean'))
fig.add_trace(go.Scatter(x=x1, y=minfreq_mean, mode='lines', name='minfreq_mean'))
fig.update_layout(
    plot_bgcolor='white',  # 배경 색상 변경
    paper_bgcolor='white',  # 텍스트 색상 변경
    font_color='#080808'  # 그리드 색상 변경
)

pyo.iplot(fig)
# plt.style.use('grayscale') 


# ax = sns.histplot(data=[mfcc_mean, spectral_centroids_mean, spectral_bandwidth_mean, spectral_contrast_mean, spectral_rolloff_mean, meanfreq_mean, maxfreq_mean, minfreq_mean], 
#                   kde=True, stat="density", common_norm=False)
# ax.set_xlim(-1000, 5000)
```
